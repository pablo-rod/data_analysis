# Most popular video game ranked by specialists in IGN website

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

Exploring data using the "Platform" column

```python
scores_data = "/content/ign_scores.csv"
scores_file = pd.read_csv(scores_data, index_col= "Platform")
scores_file
```

Highest Scores per genre

```python
scores_file.max()
```

## Bar chart showing average score for simulation games by platform




```python
plt.figure(figsize = (20, 7))
plt.title('Average score for simulation games by platform')
sns.barplot(x=scores_file.index, y = scores_file['Simulation'])
```

## Analising using a heatmap

```python
plt.figure(figsize = (20, 7))
plt.title("Heatmap showing average game score by platform and genre")
sns.heatmap(data=scores_file, annot=True)
```

# **PlayStation 4** is the platform that best performs for **Simulation** games (9.2). The **Game boy color** platform has the worse performance with **Shooter** and **Fighting** games (4.5).
