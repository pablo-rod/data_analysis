```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
candy_data = pd.read_csv('/content/candy.csv')
```

```python
candy_data.head()
```

### Do people tend to prefer candies with higher sugar content?

```python
sns.scatterplot(x = candy_data['sugarpercent'], y = candy_data['winpercent'])
```

### Creating the same scatter plot created before, but now with a regression line.

```python
sns.regplot(x = candy_data['sugarpercent'], y = candy_data['winpercent'])
```

### Scatter plot showing the relationship between 'pricepercent', 'winpercent', and 'chocolate'

```python
sns.scatterplot(x=candy_data['pricepercent'], y=candy_data['winpercent'], hue=candy_data['chocolate'])
```

### Creating the same scatter plot you created before, but now with two regression lines, corresponding to (1) chocolate candies and (2) candies without chocolate.

```python
sns.lmplot(x='pricepercent', y='winpercent', hue='chocolate', data=candy_data)
```

```python
sns.lmplot(x='pricepercent', y='winpercent', data=candy_data)
```

### Creating a categorical scatter plot to highlight the relationship between 'chocolate' and 'winpercent'.

```python
sns.swarmplot(x=candy_data['chocolate'],
              y=candy_data['winpercent'])
```
