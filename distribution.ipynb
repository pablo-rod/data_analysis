# Working with a real-world dataset containing information collected from microscopic images of breast cancer tumors.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
cancer_data = pd.read_csv('/content/cancer.csv', index_col="Id")
```

```python
cancer_data.head()
```

Creating two histograms that show the distribution in values for 'Area (mean)', separately for both benign and malignant tumors.

```python
sns.histplot(data=cancer_data, x='Area (mean)', hue='Diagnosis')
```

Creating two KDE plots that show the distribution in values for 'Radius (worst)', separately for both benign and malignant tumors.

```python
sns.jointplot(x=cancer_data['Radius (worst)'], y=cancer_data['Diagnosis'])
```
